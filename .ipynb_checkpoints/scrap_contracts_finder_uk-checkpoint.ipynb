```python
from time import sleep
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import  WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options as Options
from selenium.webdriver.common.keys import Keys

chrome_options = Options()

# executable_path is the path where yu store your webdriver
driver = webdriver.Chrome(executable_path = r'chromedriver', options=chrome_options)

contract_urls = [] #url of every poststore here
all_data = []

driver.get("https://www.contractsfinder.service.gov.uk/Search/Results?page=1")
    
# path to chrome driver C:\Program Files\ChromeDriver  
sleep(7)
```

```python
"""Foreign Commonwealth and Development Office"""
# we use the tag name and the class name of the link
urls = driver.find_elements(By.XPATH, "//a[@class = 'govuk-link search-result-rwh break-word']")

#convert to urls
for link in urls:
    all_links = link.get_attribute("href")
    contract_urls.append(all_links)
```

```python
# we get all the url for every post on the website
contract_urls
```

```python
"""we go on every of the link and scrap it's data"""
for i in contract_urls:
    driver.get(i) #going to every particular page and extracting links one by one
    sleep(5)
    #extract data now. all data are found in the tag div, with class name content-block
    contract_details = driver.find_element(By.XPATH, "//div[@class = 'content-block']")
    contract_text = contract_details.text
    #split the text by new line
    data = contract_text.split("\n")
    all_data.append(data)

# convert the all_data list into a dataframe
df = pd.DataFrame(all_data)
print(df)
```

```python
# Save scrapped data as an excel file
df.to_csv("Commonwealth_data2.csv", index=False)
```
